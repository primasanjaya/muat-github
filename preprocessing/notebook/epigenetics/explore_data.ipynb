```python
import numpy as np
import os 
import pandas as pd
```

```python
data_dir = '/csc/epitkane/projects/EpicSV/data/master_bedgraphs/'
```

```python
filename = 'master_ChromHMM_145_epigenomes_hg19.bedg.gz'
```

```python
data = pd.read_csv(data_dir + filename, sep='\t')
```

```python
data
```

```python
simplest = data.iloc[:,3:]
```

```python
list_col = simplest.columns
```

```python
allset = set()

for i in list_col:
    onelist = simplest[i].to_list()
    onelist = set(onelist)
    allset = allset.union(onelist)
```
